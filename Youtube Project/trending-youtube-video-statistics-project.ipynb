```python
likes_list=list(df['likes'])
dislikes_list=list(df['dislikes'])
```

```python
def likes_dislikes_rate(dataset):
    # Extract lists from dataset
    likes_list = list(dataset['likes'])
    dislikes_list = list(dataset['dislikes'])
    
    # Zip the likes and dislikes into pairs
    liste = list(zip(likes_list, dislikes_list))
    
    # Initialize an empty list to store rates
    rate_list = []
    
    # Iterate over the likes and dislikes pairs
    for like, dislike in liste:
        # Check if the sum of likes and dislikes is zero to avoid division by zero
        if (like + dislike) == 0:
            rate_list.append(0)
        else:
            rate_list.append(like / (like + dislike))
    
    return rate_list

# Example call to the function
df['likes_rates']=likes_dislikes_rate(df)
df.head()
```

```python
df.sort_values('likes_rates', ascending=False)[['title','likes','dislikes','likes_rates']]
```
