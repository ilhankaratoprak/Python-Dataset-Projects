```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
df=pd.read_csv('/kaggle/input/starbucks-nutrition/starbucks.csv',index_col=0)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df['item'].nunique()
```

```python
df['item'].unique()
```

```python
df['type'].unique()
```

```python
df.groupby('type')['item'].count()
```

```python
df.groupby('type')['item'].count().plot()
plt.title('Type of Product')
plt.show()
```

```python
sns.countplot(x='type',data=df, palette='Set1')
plt.title('Count of Product')
plt.show()
```

```python
sns.catplot(kind='bar', x='type',y='protein',data=df)
plt.title('Protein by Type')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot(kind='bar', x='type',y='calories',data=df)
plt.title('Calories by Type')
plt.xticks(fontsize=7, color='blue')
plt.show()
```

```python
sns.catplot(kind='bar', x='type',y='fiber',data=df)
plt.title('Fiber by Type', color='red')
plt.xticks(fontsize=7)
plt.show()
```

```python
df_corr=df[['calories', 'fat', 'carb', 'fiber', 'protein']].corr()
df_corr
```

```python
sns.heatmap(df_corr.corr(),annot=True)
```

```python
plt.title('Caleries and Fat')
sns.scatterplot(x='calories',y='fat', data=df,color='yellow',edgecolor='red')
plt.plot()
```

```python
sns.displot(x='calories',data=df,color='red',kde=True)
plt.title('Calorie  Chart')
plt.show()
```
